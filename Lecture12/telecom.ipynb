```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop(["Churn", "State"], axis=1)
X["International plan"] = X["International plan"].map({"Yes": 1, "No": 0})
X["Voice mail plan"] = X["Voice mail plan"].map({"Yes": 1, "No": 0})
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
tsne = TSNE(random_state=17)
tsne_repr = tsne.fit_transform(X_scaled)
```

```python
plt.scatter(tsne_repr[:, 0], tsne_repr[:, 1], alpha=0.5);
```

We can see the data has six clusters in lower dimensional feature space of different size and concentration. High concentration of target cancelling in the cluster lower left

```python
plt.scatter(
    tsne_repr[:, 0],
    tsne_repr[:, 1],
    c=df["Churn"].map({False: "blue", True: "orange"}),
    alpha=0.5,
);
```

```python
_, axes = plt.subplots(1, 2, sharey=True, figsize=(12, 5))

for i, name in enumerate(["International plan", "Voice mail plan"]):
    axes[i].scatter(
        tsne_repr[:, 0],
        tsne_repr[:, 1],
        c=df[name].map({"Yes": "orange", "No": "blue"}),
        alpha=0.5,
    )
    axes[i].set_title(name);
```

From the plot we can see that many dissatisfied customers who canceled their subscription are crowded together in one cluster representing the people with the international plan but no voice mail.
